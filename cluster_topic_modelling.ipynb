```python
import re
import os
import time
import numpy as np
import pandas as pd
from pprint import pprint
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
from gensim.models import LdaMulticore
from gensim.corpora import Dictionary
```

# 5 clusters - k-means

```python
cluster = [f for f in os.listdir('temp/Lemmatized/lemm/') if f.startswith('lenta')]
```

```python
print(cluster)
```

```python
%%time
docs = []
for i in cluster:
    with open('temp/Lemmatized/lemm/{}'.format(i), 'r', encoding='UTF-8') as f:
        docs.append(f.readlines())
        txts = []
for d in docs:
    for dd in d:
        txts.append(dd.split())
```

```python
len(txts)
```

```python
%%time
dictionary = Dictionary(txts)
dictionary.filter_extremes(no_below=1000, no_above=0.9)
corpus = [dictionary.doc2bow(txt) for txt in txts]
```

```python
%%time
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                     id2word=dictionary,
                        num_topics=13, 
                        chunksize=1000,
                        passes=50)
```

```python
lda_model.print_topics()
```

```python
lda_model.save('model_5cl_1.model')
```
